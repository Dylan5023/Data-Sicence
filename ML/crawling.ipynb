```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.hollys.co.kr/store/korea/korStore2.do'
print(url)
```

```python
sess = requests.Session()

response = sess.get(url)

html = response.text
soup = BeautifulSoup(html, 'html.parser')
```

```python
print(soup.prettify())
```

```python

tag_tbody = soup.find('tbody')
print(tag_tbody.prettify())
```

```python
tag_tr = tag_tbody.find_all('tr')

print(len(tag_tr))
```

```python
results = []

for store in tag_tr:
    store_td = store.find_all('td')
    store_location = store_td[0].string
    store_name = store_td[1].string
    store_address = store_td[3].string
    store_phone = store_td[5].string
    results.append([store_name, store_location, store_address, store_phone])
```

```python


page = 56
url = 'https://www.hollys.co.kr/store/korea/korStore2.do?pageNo=%d&sido=&gugun=&store=' %page
print(url)
```

```python
results = []
for page in range(1, 57):
    url = 'https://www.hollys.co.kr/store/korea/korStore2.do?pageNo=%d&sido=&gugun=&store=' %page
    print(url)

    sess = requests.Session()
    response = sess.get(url)
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')

    tag_tbody = soup.find('tbody')
    tag_tr = tag_tbody.find_all('tr')


    for store in tag_tr:
        store_td = store.find_all('td')
        store_location = store_td[0].string
        store_name = store_td[1].string
        store_address = store_td[3].string
        store_phone = store_td[5].string
        results.append([store_name, store_location, store_address, store_phone])



```

```python
import pandas as pd

hollys_df = pd.DataFrame(results, columns=['매장명', '지역', '주소', '전화번호'])
hollys_df
```
