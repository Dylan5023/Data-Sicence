# one-hot-encoding

```python
df.head()
```

```python
blood_map = {
    'A' : 0,
    'B': 1,
    'AB': 2,
    '0': 3

}
```

```python
df['혈액형_code'] = df['혈액형'].map(blood_map)
```

```python
df.head()
```

```python
df['혈액형_code'].value_counts()
```

```python
df['혈액형_code']
```

```python
pd.get_dummies(df['혈액형_code'])
```

```python
#use prefix 
pd.get_dummies(df['혈액형_code'], prefix = '혈액형')
```
