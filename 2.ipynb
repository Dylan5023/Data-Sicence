# Remove duplicate values(drop_duplicates)

```python
df = pd.read_csv('https://bit.ly/ds-korean-idol')
```

```python
df
```

```python
df['키'].drop_duplicates()
```

**keep 옵션으로 유지하고 싶은 데이터를 선택할 수 있다. keep = first/last**

```python
df['키'] = df['키'].drop_duplicates(keep = 'last')
```

```python
df
```

**행 전체 제거**

```python
df1 = df.copy()
```

```python
df
```

```python
df1.drop_duplicates('그룹')
```
